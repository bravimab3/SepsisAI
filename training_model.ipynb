```python
drop_cols = [

    # --------------------------
    # 1. IDENTIFIERS (NO PREDICTIVE VALUE)
    # --------------------------
    'Unnamed: 0',     # just row index, useless for ML
    # --------------------------
    # 2. ADMINISTRATIVE ICU INFO (NOT MEDICAL SIGNAL)
    # --------------------------
    'Unit1',
    'Unit2',      
    # 3. EXTREMELY HIGH MISSING (>90%) LAB FEATURES
    # (too sparse → imputation becomes meaningless noise)
    # --------------------------
    'EtCO2',              # 100% missing
    'TroponinI',         # ~99% missing
    'Bilirubin_direct',  # ~99% missing
    'Fibrinogen',        # ~99% missing
    'Bilirubin_total',   # ~98% missing
    'Alkalinephos',      # ~98% missing
    'AST',               # ~98% missing
    'Lactate',           # ~96% missing
    'PTT',               # ~95% missing
    'SaO2',              # ~95% missing
    'Calcium',           # ~95% missing
    'Phosphate',         # ~94% missing
    'Platelets',         # ~93% missing
    'Creatinine',        # ~93% missing
    'WBC',               # ~92% missing
    'Magnesium',         # ~92% missing
    'HCO3',              # ~91% missing
    'BUN',               # ~91% missing
    'Chloride',          # ~91% missing
    'PaCO2',             # ~91% missing
    'Hgb'                # ~91% missing
]

df = df.drop(columns=drop_cols, errors='ignore')
```

```python
#data cleaning
```

```python
# Sort by patient and time
df = df.sort_values(['Patient_ID', 'Hour'])

# Fill within each patient
vitals = ['HR','O2Sat','Temp','SBP','MAP','DBP','Resp']

df[vitals] = (
    df.groupby('Patient_ID')[vitals]
      .transform(lambda x: x.ffill().bfill())
)

# Fill remaining gaps
for col in vitals:
    df[col] = df[col].fillna(df[col].median())
```

```python
df = df.dropna(
    subset=['Age','Gender','HospAdmTime','ICULOS','SepsisLabel','Patient_ID']
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cols = ['HR','O2Sat','Temp','SBP','MAP','DBP','Resp','Age']

plt.figure(figsize=(15,10))

for i, col in enumerate(cols):
    plt.subplot(3,3,i+1)
    sns.boxplot(y=df[col])
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
Q1 = df['HR'].quantile(0.25)
Q3 = df['HR'].quantile(0.75)

IQR = Q3-Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

df['HR'] = df['HR'].clip(lower, upper)
```

```python
df = df[(df['Temp'] > 30) & (df['Temp'] < 45)]

df = df[(df['Age'] >= 0) & (df['Age'] <= 100)]
```

```python
groups = df['Patient_ID']
#target and dependent variable separate
X = df.drop(['SepsisLabel','Patient_ID'], axis=1)

y = df['SepsisLabel']
```

```python
from sklearn.model_selection import GroupShuffleSplit


gss = GroupShuffleSplit(
    test_size=0.2,
    random_state=42
)

train_idx, test_idx = next(
    gss.split(X, y, groups)
)


X_train = X.iloc[train_idx]
X_test = X.iloc[test_idx]

y_train = y.iloc[train_idx]
y_test = y.iloc[test_idx]
```

```python
from sklearn.preprocessing import RobustScaler


scaler = RobustScaler()


X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)
```

```python
#feature engineering
```

```python
X_train['Shock_Index'] = X_train['HR'] / X_train['SBP']

X_test['Shock_Index'] = X_test['HR'] / X_test['SBP']

X_train['ICULOS'] = np.log1p(X_train['ICULOS'])
X_test['ICULOS'] = np.log1p(X_test['ICULOS'])

X_train['HospAdmTime'] = np.sign(X_train['HospAdmTime']) * np.log1p(abs(X_train['HospAdmTime']))

X_test['HospAdmTime'] = np.sign(X_test['HospAdmTime']) * np.log1p(abs(X_test['HospAdmTime']))
```
